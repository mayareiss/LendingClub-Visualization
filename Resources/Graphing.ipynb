```python
status_sum = df.groupby(['loan_status']).sum()['loan_amnt']
status_count = df.groupby(['loan_status']).count()['loan_amnt']
status_ave = df.groupby(['loan_status']).mean()['loan_amnt']

status_df = pd.DataFrame({"Total Loan value": status_sum,
                        "Number of Loans": status_count,
                          "Ave Loan value": status_ave
                               })

status_df = status_df.sort_values("Total Loan value")

status_df["Total Loan value"] = status_df["Total Loan value"].map("${:,.2f}".format)
status_df["Ave Loan value"] = status_df["Ave Loan value"].map("${:,.2f}".format)

status_df
```

```python
grade_sum = df.groupby(['grade']).sum()['loan_amnt']
grade_count = df.groupby(['grade']).count()['loan_amnt']
grade_ave = df.groupby(['grade']).mean()['loan_amnt']

grade_df = pd.DataFrame({"Total Loan value by Grade": grade_sum,
                        "Loans by Grade": grade_count, 
                         "Ave Loan Value": grade_ave
                               })

grade_df["Total Loan value by Grade"] = grade_df["Total Loan value by Grade"].map("${:,.2f}".format)
grade_df["Ave Loan Value"] = grade_df["Ave Loan Value"].map("${:,.2f}".format)

grade_df
```
